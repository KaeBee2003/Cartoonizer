```python
import cv2
import numpy as np
import os
```

```python
def read_img(filename):
    img = cv2.imread(filename)
    return img
```

```python
def color_quantisation(img, k):
    data = np.float32(img).reshape((-1, 3))
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 20, 0.001)
    _, label, center = cv2.kmeans(data, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    center = np.uint8(center)
    result = center[label.flatten()]
    result = result.reshape(img.shape)
    return result
```

```python
# Folder paths
input_folder = r'D:\Cartoonizer'
output_folder = r'D:\Cartoonizer\Output'
os.makedirs(output_folder, exist_ok=True)
```

```python
# Parameters
totalColors = 15
```

```python
# Process each image in the folder
for filename in os.listdir(input_folder):
    if filename.endswith(('.jpg', '.jpeg', '.png')):  # Check for image file formats
        filepath = os.path.join(input_folder, filename)
        img = read_img(filepath)
        
        # Apply color quantization
        cartoon = color_quantisation(img, totalColors)
        
        # Save the output
        output_path = os.path.join(output_folder, f"cartoon_{filename}")
        cv2.imwrite(output_path, cartoon)
```
